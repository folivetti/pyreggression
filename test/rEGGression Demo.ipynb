```python
from pyreggression import PyReggression
import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
reg = PyReggression("nasa_battery_1_10min.csv", "", "MSE", "nasa_battery.egraph", "", False)
```

```python
# Let's get the top-10
reg.top(10)
```

```python
# Now let's retrieve the top 10 with size < 10 and not having the pattern v0^v1
reg.top(10, filters=["size < 10"], pattern="v0 ** v1", negate=True)
```

```python
# Now let's find expressions with this pattern at the root (so the root should be ^)
reg.top(10, filters=["size < 10"], pattern="v0 ** v1", isRoot=True)
```

```python
# let's test other filters
reg.top(10, filters=["size < 10", "parameters > 1", "cost < 15"], criteria="by dl")
```

```python
# query for the top patterns by the following criteria
# from the top 5000 expressions, with at least 1000 of frequency
# show the top 25 distributions sorted by avg. fitness 
# and with size between 8 and 15
reg.distribution(filters=["size < 15"], limitedAt=25, dsc=True, byFitness=True, atLeast=1000, fromTop=5000)
```

```python
reg.countPattern("v0 + v1")
```

```python
# (re) optimize the parameters
reg.optimize(73451)
```

```python
# inspect the subtrees
reg.subtrees(73451)
```

```python
# insert a new expression
reg.insert("t0 + log(x0)")
```

```python
# show the Pareto front
pareto = reg.pareto(byFitness=False)
pareto
```

```python
# plot the pareto front
plt.plot(pareto.DL.iloc[3:], pareto.Size.iloc[3:], '.')
```

```python
reg.save("newegg.egg");
```
